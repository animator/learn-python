Question: Factorial: Compute the factorial of a non-negative integer n

```python
def factorial(n):
    if n<0:
        return "Invalid number"
    if n==0:
        return 1
    return n* factorial(n-1) #4! = 4*3! 
```

```python
print(factorial(1))
```

Question: Find the nth fibonacci number

```python
def fibonacci(n):
    if n==0:
        return 0
    if n==1:
        return 0
    elif n==2:
        return 1
    return fibonacci(n-1)+fibonacci(n-2)# fib(n)=fib(n-1)+fib(n-2)
```

```python
print(fibonacci(3))
```

Question: Binary Search using Recursion

```python
def binary(arr,target,low,high):
    if low>high:#target is not present in the array
        return -1
    mid =(low+high)//2
    if arr[mid]==target:
        return mid
    elif target<arr[mid]:
        return binary(arr,target,low,mid-1)
    else:
        return binary(arr,target,mid+1,high)
```

```python
arr=[1,2,3,4,5,6]
target=3
print(binary(arr,target,0,len(arr)-1))
```

Question: Power of Four using recursion

```python

def powerOfFour(number,remainder):
    if number==1 and remainder==0:
        return True
    if remainder>0:
        return False
    remainder = number%4
    number = number//4
    bool=powerOfFour(number,remainder)
    return bool
    
    
```

# Recursion Problems in Binary Trees

```python
class Node:
    """ The class represents a single node in a binary tree """
    def __init__(self, value, left_child:'Node' = None, right_child:'Node' = None):
        self.value = value
        self.left_child = left_child
        self.right_child = right_child
```

```python
#sample tree structure
tree = Node(2)
tree.left_child = Node(3)
tree.left_child.left_child = Node(5)
tree.left_child.right_child = Node(8)
tree.right_child = Node(4)
tree.right_child.right_child = Node(11)
```

Question: Inorder, Preorder and Postorder Traversal of a binary tree

```python
def inorder(root):
    if root is None:
        return 
    inorder(root.left_child)
    print(root.value)
    inorder(root.right_child)
    
def preorder(root):
    if root is None:
        return
    print(root.value)
    preorder(root.left_child)
    preorder(root.right_child)
    
def postorder(root):
    if root is None:
        return
    postorder(root.left_child)
    postorder(root.right_child)
    print(root.value)
```

```python
inorder(tree)
```

```python
preorder(tree)
```

```python
postorder(tree)
```

Question:Given the root of a binary tree and an integer targetSum, return true if the tree has a root-to-leaf path such that adding up all the values along the path equals targetSum.


Input: root = [5,4,8,11,null,13,4,7,2,null,null,null,1], targetSum = 22
Output: true
Explanation: The root-to-leaf path with the target sum is shown.

```python
def PathSum(root,target,sum):
    if root is None:
        return False
    if root.left_child is None and root.right_child is None:
        sum =sum+root.value
        if sum==target:
            return True
        
    return PathSum(root.left_child,target,sum+root.value) or PathSum(root.right_child,target,sum+root.value)
    
```

Question:Given a binary tree root and an integer target, delete all the leaf nodes with value target.

Note that once you delete a leaf node with value target, if its parent node becomes a leaf node and has the value target, it should also be deleted (you need to continue doing that until you cannot).


```python
def removeLeafNodes(root,target):
    if root is None:
        return None
    root.left_child = removeLeafNodes(root.left_child,target)
    root.right_child = removeLeafNodes(root.right_child,target)
    if root.left_child is None and root.right_child is None:
        if root.value == target:
            return None
    return root
```

```python
root= Node(1)
root.left_child = Node(2)
root.left_child.left_child = Node(2)
root.right_child = Node(3)
root.right_child.right_child = Node(4)
root.right_child.left_child = Node(2)
res = removeLeafNodes(root,2)
inorder(res)
```

 Question:Invert the binary tree
 

```python
def invert(root):
    if root is None:
        return None
    root.left_child,root.right_child = invert(root.right_child),invert(root.left_child)
    return root
```

```python
root = Node(4)
root.left_child = Node(3)
root.right_child = Node(5)
inorder(root)
invert(root)
print('After inversion')
inorder(root)
```

Question:Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

 
 
 Input: root = [3,9,20,null,null,15,7]
Output: 3

```python
def maxDepth(root):
    if root is None:
        return 0
    leftMax = maxDepth(root.left_child)+1
    rightMax = maxDepth(root.right_child)+1
    
    return max(leftMax,rightMax)
```

```python
root = Node(3)
root.left_child = Node(9)
root.right_child = Node(20)
root.right_child.left_child = Node(15)
root.right_child.right_child = Node(7)
print(maxDepth(root))
```
