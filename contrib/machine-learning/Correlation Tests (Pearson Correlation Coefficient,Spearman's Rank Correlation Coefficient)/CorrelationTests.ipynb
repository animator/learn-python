```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
data = pd.read_csv('/content/drive/MyDrive/GirlsScriptOpenSource/Mlrepo/CorrelationTests/correlation_data.csv')
```

```python
# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Perform Pearson Correlation Test
pearson_corr, pearson_p_value = pearsonr(data['x'], data['y'])
print(f"Pearson Correlation Coefficient: {pearson_corr}")
print(f"p_value: {pearson_p_value}")
```

```python
# Interpret the results
alpha = 0.05
if pearson_p_value>= alpha:
    print(f"Since p-value ({pearson_p_value:.4f}) >= alpha ({alpha}), we do not reject the null hypothesis.")
else:
    print(f"Since p-value ({pearson_p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
```

```python
# Perform Spearman Correlation Test
spearman_corr, spearman_p_value = spearmanr(data['x'], data['y'])
print(f"Spearman's Rank Correlation Coefficient: {spearman_corr}")
print(f"P-value: {spearman_p_value}")
```

```python
# Interpret the results
alpha = 0.05
if spearman_p_value>= alpha:
    print(f"Since p-value ({spearman_p_value:.4f}) >= alpha ({alpha}), we do not reject the null hypothesis.")
else:
    print(f"Since p-value ({spearman_p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
```

```python
# Plotting the data
plt.figure(figsize=(12, 6))

# Scatter plot with regression line
sns.regplot(x='x', y='y', data=data)
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
